```python
import pandas as pd
df = pd.read_excel("Maritime_Operations.xlsx", sheet_name=0)
print(df.head(5))
```

```python
df['Delay_Days'] = df['Actual_Turnaround_Days'].astype(float) - df['Planned_Turnaround_Days'].astype(float)
print(df[['Actual_Turnaround_Days', 'Planned_Turnaround_Days', 'Delay_Days']].head())
```

```python
print(df['Delay_Days'].dtype)
```

```python
df['Delay_Status'] = df['Delay_Days'].apply(lambda x: 'Late' if x > 0 else 'On Time')
```

```python
df[['Delay_Days', 'Delay_Status']].head()
```

```python
total_vessels = df.shape[0]
total_vessels
```

```python
status_counts = df['Delay_Status'].value_counts()
status_counts
```

```python
late_vessels = status_counts.get('Late', 0)
on_time_vessels = status_counts.get('On Time', 0)

late_vessels, on_time_vessels
```

```python
on_time_performance = (on_time_vessels / total_vessels) * 100
on_time_performance
```

```python
kpi_summary = pd.DataFrame({
'Metric': [
'Total Vessels',
'On-Time Vessels',
'Late Vessels',
'On-Time Performance (%)'
],
'Value': [
total_vessels,
on_time_vessels,
late_vessels,
round(on_time_performance, 2)
]
})
kpi_summary
```

```python
status_counts = df['Delay_Status'].value_counts()
status_counts
```

```python
import matplotlib.pyplot as plt

status_counts.plot(kind='bar')
plt.title('Vessel On-Time Performance')
plt.xlabel('Delay Status')
plt.ylabel('Number of Vessels')
plt.show()
```

```python
port_delay = df.groupby('Port')['Delay_Days'].mean().sort_values(ascending=False)
port_delay
```

```python
import matplotlib.pyplot as plt

port_delay.plot(kind='bar')
plt.title('Average Delay by Port')
plt.xlabel('Port')
plt.ylabel('Average Delay (Days)')
plt.show()
```

```python
port_risk = port_delay.reset_index()
port_risk.columns = ['Port', 'Avg_Delay_Days']
port_risk
```

```python
port_risk['Risk_Level'] = port_risk['Avg_Delay_Days'].apply(
lambda x: 'High Risk' if x > 0 else 'Normal'
)
port_risk
```

```python
df = df.merge(
port_risk,
left_on='Port',
right_on='Port',
how='left'
)

df.head()
```

```python
df[['Vessel_ID', 'Port', 'Delay_Days', 'Risk_Level']].head(21)
```

```python
total_vessels = df['Vessel_ID'].nunique()
total_vessels
```

```python
delayed_vessels = df[df['Delay_Status'] == 'Late']['Vessel_ID'].nunique()
delayed_vessels
```

```python
on_time_performance = ((total_vessels - delayed_vessels) / total_vessels) * 100
on_time_performance
```

```python
df['Risk_Level'].value_counts()
```

```python
def classify_risk(delay):
    if delay >= 3:
        return 'High'
    elif delay >= 1:
        return 'Medium'
    else:
        return 'Low'
```

```python
df['Risk_Level'] = df['Delay_Days'].apply(classify_risk)
```

```python
df[['Delay_Days', 'Risk_Level']]
```

```python
high_risk_vessels = df[df['Risk_Level'] == 'High']['Vessel_ID'].nunique()
high_risk_vessels
```

```python
high_risk_percentage = (high_risk_vessels / total_vessels) * 100
high_risk_percentage
```

```python
kpi_summary = pd.DataFrame({
'Metric': [
'Total Vessels',
'Delayed Vessels',
'On-Time Performance (%)',
'High-Risk Vessels',
'High-Risk Exposure (%)'
],
'Value': [
total_vessels,
delayed_vessels,
round(on_time_performance, 2),
high_risk_vessels,
round(high_risk_percentage, 2)
]
})

kpi_summary
```
